```python
seenPairs = []
pairs = []
matrix = [[1, 2],[3, 4]]
letter1 = ""
letter2 = ""
combined = ""
# goes over every pair of letters
for i in range(26):
    for j in range(26):
        # applies the matrix multiplication to the combinations of letters
        letter1 = chr((((matrix[0][0] * i) + (matrix[0][1] * j)) % 26) + 97)
        letter2 = chr((((matrix[1][0] * i) + (matrix[1][1] * j)) % 26) + 97)
        combined = (letter1 + letter2, f"({i}, {j})")
        # checks if the combination had been seen before
        for k in range(len(seenPairs)):
            if(combined[0] == seenPairs[k][0]):
                matched = seenPairs[k]
                print(matched[0] + matched[1] + " | " + combined[0], combined[1])
                pairs.append((matched[0] + matched[1] + " | " + combined[0], combined[1]))
        # if it hasnt been seen before then it adds it to memory for checking later
        else:
            seenPairs.append(combined)
        # print(letter1+letter2)
```
